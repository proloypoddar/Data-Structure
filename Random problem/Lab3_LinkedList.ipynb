```python
# Run This cell Second
class Node:
  def __init__(self, e, n):
    self.element = e
    self.next = n
```

```python
# To you coding in this cell
class LinkedList:
  
  def __init__(self, a):

    if type(a)==list:
      self.head=Node(a[0],None)
      curr=self.head
      for i in range(1,len(a)):
        curr.next=Node(a[i],None)
        curr=curr.next
    else:
      self.head=a
  def traverseList(self):
    s = ''
    temp = self.head
    while temp != None:
      if temp.next != None:
        s += str(temp.element) + " "
      else:
        s += str(temp.element)
      temp = temp.next
    return s
  def countNode(self):
    s = ''
    c=0
    temp = self.head
    while temp != None:
      c+=1
      temp = temp.next
    return c

  def nodeAt(self, idx):
    c=0
    temp= self.head
    obj=None
    while temp!= None:
      if c==idx:
        obj=temp
        break
      temp=temp.next
      c+=1
    return obj

  def get(self, idx):
    c=0
    obj=None
    temp=self.head
    while temp != None:
      if c==idx:
        return temp.element
      temp=temp.next
      c+=1
    return None


  def set(self, idx, elem):
    c=0
    temp=self.head
    while temp != None:

      if c==idx:
        old=temp.element
        temp.element= elem
        return old
      temp=temp.next
      c+=1
    return None
    

  def indexOf(self, elem):
    temp = self.head
    idx = 0
    while temp != None:
        if temp.element == elem:
            return idx
        temp = temp.next
        idx += 1
    return -1

  def contains(self, elem):
    
    temp=self.head
    while temp != None:
      if temp.element==elem:
        return True
      temp=temp.next
    return False
  def copyList(self):
    new_head=None
    new_tail=None
    temp=self.head
    while temp!= None:
      n=Node(temp.element,None)
      if new_head==None:
        new_head=n
        new_tail=new_head
      else:
        new_tail.next=n
        new_tail=new_tail.next
      temp=temp.next
    return new_head
  def reverseList(self):
    prev = None
    current = self.head
    while current != None:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    self.head = prev


  def insert(self, elem, idx):
    if idx < 0:
      print("Invalid Index")
      return
    if idx == 0:
      self.head = Node(elem, self.head)
      return
    curr = self.head
    for i in range(idx-1):
      if curr.next == None:
        print("Invalid Index")
        return
      curr = curr.next
    curr.next = Node(elem, curr.next)
  def remove(self, idx):
    if idx < 0:
      print("Invalid Index")
      return None
    if idx == 0:
      elem = self.head.element
      self.head = self.head.next
      return elem
    curr = self.head
    prev = None
    for i in range(idx):
      if curr is None:
        print("Invalid Index")
        return None
      prev = curr
      curr = curr.next
    elem = curr.element
    prev.next = curr.next
    return elem
  # Rotates the list to the left by 1 position.
  def rotateLeft(self):
    if self.head != None or self.head.next != None:
      return
      
    new_head = self.head.next
    self.head.next = None
    self.tail = self.head
    self.head = new_head
    self.tail.next = self.head
  
  # Rotates the list to the right by 1 position.
  def rotateRight(self):
    if self.head is None or self.head.next is None:
      return
    curr = self.head
    while curr.next.next is not None:
      curr = curr.next
    curr.next.next = self.head
    self.head = curr.next
    curr.next = None   
```
