```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):

    self.head = Node(a[0], None, None)
    curr = self.head
    for i in range(1, len(a)):
        new_node = Node(a[i], curr, None)
        curr.next = new_node
        new_node.prev = curr
        curr = new_node
    curr.next = self.head
    self.head.prev = curr

  
  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    temp=self.head
    temp2=temp.next
    c=1
    self.length= c
    while temp2 !=temp:
      c+=1
      temp2=temp2.next
    return c
  
  # prints the elements in the list
  def forwardprint(self):
    if self.head == None:
      return
    curr = self.head
    print(curr.element)
    curr = curr.next
    while curr != self.head:
        print(curr.element)
        curr = curr.next
  # prints the elements in the list backward
  def backwardprint(self):
    if self.head == None:
        return
    curr = self.head.prev
    print(curr.element)
    curr = curr.prev
    while curr != self.head.prev:
        print(curr.element)
        curr = curr.prev
  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    temp = self.head
    temp2 = temp.next
    c = 1
    while temp2 != temp:
        if c == idx:
            return temp2
        c += 1
        temp2 = temp2.next
    return None
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    temp = self.head
    temp2 = temp.next
    c = 1
    while temp2 != temp:
        if temp2.element == elem:
            return c
        c += 1
        temp2 = temp2.next
    return -1
  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx): 

    if idx < 0:
        print("Invalid index")
        return
    new_node = Node(elem, None, None)
    if idx == 0: # Insert at the beginning
        new_node.next = self.head
        new_node.prev = self.head.prev
        self.head.prev.next = new_node
        self.head.prev = new_node
        self.head = new_node
    else:
        curr = self.head
        for i in range(idx-1):
            if curr.next == self.head:
                print("Invalid index")
                return
            curr = curr.next
        new_node.next = curr.next
        new_node.prev = curr
        curr.next.prev = new_node
        curr.next = new_node
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):

    if idx < 0 or idx >= self.countNode():
        return None
    curr = self.head
    if idx == 0:
        self.head = self.head.next
        self.head.prev = curr.prev
    else:
        for i in range(idx):
            curr = curr.next
        curr.prev.next = curr.next
        if curr.next is not None:
            curr.next.prev = curr.prev
    removed_element = curr.element
    curr = None
    return str(removed_element)
```
